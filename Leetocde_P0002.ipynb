# Plus One

https://leetcode.com/problems/plus-one/description/

## Solution 1: Use Type Conversion

```python
def plusOne_conversion(digits):
    num_str = "".join(map(str, digits))

    num_int = int(num_str) + 1

    return [int(x) for x in str(num_int)]
print(plusOne_conversion([1, 2, 3]))
print(plusOne_conversion([4, 3, 2, 1]))
print(plusOne_conversion([9]))
```

## Solution 2: Use Loop and Carry

```python
def plusOne_loop(digits):
    for i in range(len(digits) - 1, -1, -1):
        if digits[i] < 9:
            digits[i] += 1
            return digits
        else:
            digits[i] = 0
    return [1] + digits

print(plusOne_loop([1, 2, 3]))
print(plusOne_loop([4, 3, 2, 1]))
print(plusOne_loop([9]))
```
