# Valid Parentheses

https://leetcode.com/problems/valid-parentheses/description/

## Solution 1: Use Stack and Dictionary

```python
def isValid_stack(s):
    stack = []
    mapping = {")": "(", "]": "[", "}": "{"}

    for char in s:
        if char in mapping:
            top_element = stack.pop() if stack else '#'
            if mapping[char] != top_element:
                return False
        else:
            stack.append(char)
    return not stack
print(isValid_stack("()"))
print(isValid_stack("()[]{}"))
print(isValid_stack("(]"))
```

## Solution 2: Use String Replace

```python
def isValid_replace(s):
    while "()" in s or "[]" in s or "{}" in s:
        s = s.replace("()", "")
        s = s.replace("[]", "")
        s = s.replace("{}", "")
    return s == ""

print(isValid_replace("()"))
print(isValid_replace("()[]{}"))
print(isValid_replace("(]"))
print(isValid_replace("{[]}"))
```
