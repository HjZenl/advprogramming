# Binary Search

https://leetcode.com/problems/binary-search/description/

## Solution 1: Iterative Approach

```python
def search_iterative(nums, target):
    left, right = 0, len(nums) - 1
    while left <= right:
        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1

print(search_iterative([-1, 0, 3, 5, 9, 12], 9))
print(search_iterative([-1, 0, 3, 5, 9, 12], 2))
```

## Solution 2: Recursive Approach

```python
def search_recursive(nums, target):
    def helper(left, right):
        if left > right:
            return -1

        mid = (left + right) // 2

        if nums[mid] == target:
            return mid
        elif nums[mid] < target:
            return helper(mid + 1, right)
        else:
            return helper(left, mid - 1)

    return helper(0, len(nums) - 1)

print(search_recursive([-1, 0, 3, 5, 9, 12], 9))
print(search_recursive([-1, 0, 3, 5, 9, 12], 2))
```
